## Question 1
Print the result of **A+B** and explain the mechanism behind this. 

You answer: 

```python
A = np.array([[0, 1, 2]])
B = np.array([[10, 10, 10], [20, 20, 20], [30, 30, 30]])
print(A + B)
```

Numpy arrays support [broadcasting](https://numpy.org/doc/stable/user/basics.broadcasting.html), which is used to perform array operations even if these arrays are not the same shape. To perform a broadcast operation, each array must have the same number of rows, or one of the arrays must be of size 1x1. The smaller array is then copied such that the two arrays are the same size, and then the desired operation is computed.

In this example, since array A is 1x3 and array B is 3x3, A is expanded to a 3x3 array by copying the first row three times and then added to B.

## Question 4
For the previous generated data, consider each row of the array as an individual data point, with the two columns representing its distinct features. Identify the data points with the smallest and largest sum of features. Mark them in red and blue in the figure respectively.

Hint: You can compute the sum of each row and store it in an array named "sums". Then find the minimum or maximum index of this array. 

```python
# Code from previous section
SEED = 10413015
np.random.seed(SEED) # Please use your student id as the seed number
rand_array = np.random.rand(100,2)
plt.scatter(rand_array[:,0], rand_array[:,1], color="black")
plt.title("Random data generated from student ID")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

# Init values for sum finding loop
sum_min = rand_array[0][0] + rand_array[0][1]
min_pnt = [rand_array[0][0], rand_array[0][1]]
sum_max = rand_array[0][0] + rand_array[0][1]
max_pnt = [rand_array[0][0], rand_array[0][1]]
current_sum = 0
for xvalue, yvalue in rand_array:
    current_sum = xvalue + yvalue
    if current_sum > sum_max:
        max_pnt = [xvalue, yvalue]
        sum_max = current_sum
    if current_sum < sum_min:
        min_pnt = [xvalue, yvalue]
        sum_min = current_sum
plt.scatter(max_pnt[0], max_pnt[1], color="blue")
plt.scatter(min_pnt[0], min_pnt[1], color="red")
```

To find the maximum and minimum sum of x and y values, we iterate over the array and sum each point. If the combination is greater than the max sum, replace the old max and store the points. Similarly, if the sum is less than the min sum, replace the old min and store the points.

Once the loop is complete, `plt.scatter()` is called. Since the same figure is open, `plt.scatter()` will plot to the same figure, overwritting the old points with the color-coded points.

## Question 5
Add two lines representing the mean of the Feature 1 and the 25% quantile of the Feature 2 seperately. Add the legend of the two lines.

```python
# Code from previous section
SEED = 10413015
np.random.seed(SEED) # Please use your student id as the seed number
rand_array = np.random.rand(100,2)
plt.scatter(rand_array[:,0], 
            rand_array[:,1], 
            color="black",
            label="Data")
plt.title("Random data generated from student ID")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

# Init values for sum finding loop
sum_min = rand_array[0][0] + rand_array[0][1]
min_pnt = [rand_array[0][0], rand_array[0][1]]
sum_max = rand_array[0][0] + rand_array[0][1]
max_pnt = [rand_array[0][0], rand_array[0][1]]
current_sum = 0
for xvalue, yvalue in rand_array:
    current_sum = xvalue + yvalue
    if current_sum > sum_max:
        max_pnt = [xvalue, yvalue]
        sum_max = current_sum
    if current_sum < sum_min:
        min_pnt = [xvalue, yvalue]
        sum_min = current_sum
plt.scatter(max_pnt[0], max_pnt[1], color="blue")
plt.scatter(min_pnt[0], min_pnt[1], color="red")

# Code for this question
plt.axvline(np.mean(rand_array[:,0]),
            color="green", 
            label="Mean")
plt.axhline(np.quantile(rand_array[:,1], 0.25),
            color="orange",
            label="25% Quantile")
plt.legend(loc="lower right")
```

`numpy.mean()` was used to find the mean value of the x-data in the random array. [`numpy.quantile()`](https://numpy.org/doc/stable/reference/generated/numpy.quantile.html) was used to find the 25% quantile of y-data. 

Once these values were computed, they were used to place lines on the graph.
